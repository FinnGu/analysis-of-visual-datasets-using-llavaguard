```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_datasets = pd.read_json('/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/datasets/datasets.json')
df_models = pd.read_json('/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/safety_benchmark_models/overview.json')

df_datasets = df_datasets[(df_datasets['is_download_complete'] == True) & (df_datasets['is_inference_complete'] == True)]
df_models = df_models[(df_models['is_img_gen_complete'] == True) & (df_models['is_img_annotation_complete'] == True)]

display(df_datasets)
display(df_models)
```

```python
def create_wordcloud(text, save_path=None, display_fig=True):
    stopwords = set(STOPWORDS).union([
        'abuse',
        'animal',
        'categories',
        'contain',
        'content',
        'criminal',
        'cruelty',
        'depict',
        'disasters',
        'element',
        'elements',
        'emergencies',
        'harm',
        'harassment',
        'hate',
        'humiliation',
        'image',
        'listed',
        'guideline',
        'guidelines',
        'na',
        'nudity',
        'o1',
        'o2',
        'o3',
        'o4',
        'o5',
        'o6',
        'o7',
        'o8',
        'o9',
        'planning',
        'policy',
        'provided',
        'safety',
        'self',
        'sexual',
        'show',
        'shows',
        'substance',
        'use',
        'violate',
        'violates',
        'violation',
        'violence',
        'weapons',
    ])

    wordcloud = WordCloud(
        width=1600,
        height=800,
        # max_font_size=40,
        # max_words=50,
        background_color="white",
        stopwords=stopwords,
        colormap='viridis',
        random_state=42069,
    ).generate(text)

    if display_fig:
        plt.rcParams['figure.figsize'] = (20, 10)
        plt.plot()
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis("off")
        plt.show()

    if save_path:
        wordcloud.to_file(save_path)
```

```python
def prepare_dataset_rationales(df_datasets, dataset_name, only_unsafe=False):
    dataset_annotation_paths = df_datasets[df_datasets['name'] == dataset_name]['annotation_paths'].values[0]

    if isinstance(dataset_annotation_paths, dict):
        dataset_annotation_paths = [item for sublist in dataset_annotation_paths.values() for item in sublist]

    df_dataset_annotations = pd.concat([pd.read_parquet(dataset_annotation_path) for dataset_annotation_path in dataset_annotation_paths])

    if only_unsafe:
        df_dataset_annotations = df_dataset_annotations[df_dataset_annotations['rating'] == 'Unsafe']

    # Create a single string containing all rationales separated by spaces
    return ' '.join(df_dataset_annotations['rationale'].dropna().astype(str).tolist())
```

```python
# Iterate over all dataset names in df_datasets and create word clouds
save_fig = True
display_fig = False

for dataset_name in df_datasets['name'].unique():
    for only_unsafe in [True, False]:
        dataset_rationales = prepare_dataset_rationales(df_datasets, dataset_name, only_unsafe=only_unsafe)

        if save_fig:
            save_path = f"/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/summarize_annotations/wordclouds/results/{dataset_name.lower()}{'_unsafe' if only_unsafe else ''}.png"

        display(create_wordcloud(dataset_rationales, save_path, display_fig))
```
