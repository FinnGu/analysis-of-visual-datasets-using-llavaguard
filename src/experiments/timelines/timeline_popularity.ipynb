```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df_datasets = pd.read_json('/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/datasets/datasets.json')
df_models = pd.read_json('/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/safety_benchmark_models/overview.json')

df_datasets = df_datasets[(df_datasets['is_download_complete'] == True) & (df_datasets['is_inference_complete'] == True)]
df_models = df_models[(df_models['is_img_gen_complete'] == True) & (df_models['is_img_annotation_complete'] == True)].sort_values(by=['release_date'])

display(df_datasets)
display(df_models)
```

```python
def get_unsafe_percentage(df: pd.DataFrame, name_model_or_dataset: str = None) -> pd.DataFrame | float:
    def get_unsafe_percentage_for_single_entry(df, name_model_or_dataset):
        annotation_paths = df[df['name'] == name_model_or_dataset]['annotation_paths'].values[0]

        if isinstance(annotation_paths, dict):
            annotation_paths = [item for sublist in annotation_paths.values() for item in sublist]

        # We are only interested in counting 'Unsafe' ratings
        df_annotations = pd.concat([pd.read_parquet(annotation_path, columns=['rating']) for annotation_path in annotation_paths])
        total_count = len(df_annotations.index)

        # Count again after filtering
        df_annotations = df_annotations[df_annotations['rating'] == 'Unsafe']
        return len(df_annotations.index) / total_count * 100

    if name_model_or_dataset:
        return get_unsafe_percentage_for_single_entry(df, name_model_or_dataset)

    df['unsafe_percentage'] = [get_unsafe_percentage_for_single_entry(df, name_model_or_dataset) for name_model_or_dataset in df['name']]
    return df
```

```python
mode = 'datasets'

if mode == 'datasets':
    df = df_datasets
    title = "Popularity of Public Visual Datasets"
else:
    df = df_models
    title = "Popularity of Public Generative Models"

# Calculate and transform additional columns
df = get_unsafe_percentage(df)
df['release_date'] = pd.to_datetime(df['release_date'])

plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

ax = sns.scatterplot(
    data=df, 
    x='release_date', 
    y='unsafe_percentage', 
    size='downloads_last_month' if mode == 'models' else 'citations_in_2025',
    sizes=(50, 1000),
    alpha=0.7,
    hue='name',
    legend='brief',
)
plt.title(title)
plt.xlabel('Release Date')
plt.ylabel('% Unsafe Detections')

if mode == 'datasets':
    args_legend = dict(
        bbox_to_anchor=(0.5, -0.35),  # Centered below plot, adjust -0.35 as needed
        ncol=min(len(df['name'].unique()), 5),  # Use up to 5 columns
    )

    # Configure lineplot for smoothed_unsafe percentage
    df['smoothed_unsafe_percentage'] = df['unsafe_percentage'].transform(lambda x: x.rolling(window=1, min_periods=1).mean())
    annotation_anchor = df.sort_values('release_date').iloc[3]

    args_annotation = dict(
        text='Unsafe Detections',
        xy=(annotation_anchor['release_date'], annotation_anchor['smoothed_unsafe_percentage']),
        xytext=(-60, 30),
    )
else:
    args_legend = dict(
        bbox_to_anchor=(0.5, -0.5),  # Centered below plot, adjust -0.5 as needed
        ncol=min(len(df['name'].unique()), 2),  # Use up to 2 columns
    )

    # Configure lineplot for smoothed_unsafe percentage
    df['smoothed_unsafe_percentage'] = df['unsafe_percentage'].transform(lambda x: x.rolling(window=4, min_periods=1).mean())
    annotation_anchor = df.sort_values('release_date').iloc[6]

    args_annotation = dict(
        text='Smoothed Unsafe Detections',
        xy=(annotation_anchor['release_date'], annotation_anchor['smoothed_unsafe_percentage']),
        xytext=(-55, 30),
    )

# Add a line to the scatterplot, showing the smoothed unsafe_percentage
sns.lineplot(
    data=df, 
    x='release_date', 
    y='smoothed_unsafe_percentage',
    linestyle='--', 
    color='tab:blue'
    # label='Total'
)

# Annotate the lineplot
plt.annotate(
    textcoords='offset points',
    va='center',
    fontsize=12,
    color='tab:blue',
    arrowprops=dict(arrowstyle='-', color='tab:blue'),
    **args_annotation
)

# Add a custom legend for the 'hue' variable
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(
    handles=handles[1:len(df['name'].unique())+1], 
    labels=labels[1:len(df['name'].unique())+1], 
    title='', 
    loc='lower center',
    frameon=True,
    **args_legend,
)

# Set x-axis to show one label per year
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

# Format y-axis labels to show percentage
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x)}%'))

plt.show()
```
