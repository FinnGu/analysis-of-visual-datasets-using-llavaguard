```python
import pandas as pd
import seaborn as sns
import seaborn.objects as so
import matplotlib.pyplot as plt
```

```python
df_datasets = pd.read_json('/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/datasets/datasets.json')
df_models = pd.read_json('/pfss/mlde/workspaces/mlde_wsp_KIServiceCenter/finngu/LlavaGuard/src/experiments/safety_benchmark_models/overview.json')

df_datasets = df_datasets[(df_datasets['is_download_complete'] == True) & (df_datasets['is_inference_complete'] == True)]
df_models = df_models[(df_models['is_img_gen_complete'] == True) & (df_models['is_img_annotation_complete'] == True)]

display(df_datasets)
display(df_models)
```

```python
df = df_datasets
annotation_paths = df[df['name'] == 'LSUN']['annotation_paths'].values[0]

if isinstance(annotation_paths, dict):
    annotation_paths = [item for sublist in annotation_paths.values() for item in sublist]

df_annotations = pd.concat([pd.read_parquet(annotation_path) for annotation_path in annotation_paths])

df_unsafe_O4 = df_annotations[(df_annotations['rating'] == 'Unsafe') & (df_annotations['category'] == 'O4: Nudity Content')]
```

```python
def get_category_counts_unsafe(df, name_model_or_dataset):
    annotation_paths = df[df['name'] == name_model_or_dataset]['annotation_paths'].values[0]

    if isinstance(annotation_paths, dict):
        annotation_paths = [item for sublist in annotation_paths.values() for item in sublist]

    df_annotations = pd.concat([pd.read_parquet(annotation_path) for annotation_path in annotation_paths])

    total_count = len(df_annotations.index)

    # We are only interested in 'Unsafe' ratings
    df_annotations = df_annotations[df_annotations['rating'] == 'Unsafe']
    unsafe_percentage = len(df_annotations.index) / total_count * 100

    # Count the occurrences of each category
    category_counts = df_annotations['category'].value_counts()
    df_counts = pd.DataFrame(category_counts).reset_index()

    df_counts['percentage'] = df_counts['count'] / total_count * 100

    # Create column 'category_short' and order by it
    df_counts['category_short'] = df_counts['category'].str[:2]
    df_counts = df_counts.sort_values(by='category_short')

    df_counts = df_counts[df_counts['category_short'] != 'NA']

    return df_counts, unsafe_percentage
```

```python
# Change this to switch between datasets and models
# mode = 'datasets'
mode = 'models'

df = df_datasets if mode == 'datasets' else df_models

df_counts = pd.DataFrame()
df_totals = pd.DataFrame()

for row in df.itertuples():
    df_model_counts, unsafe_percentage = get_category_counts_unsafe(df, row.name)

    df_model_counts['name'] = row.name
    df_model_counts['release_date'] = row.release_date
    
    df_counts = pd.concat([df_counts, df_model_counts], ignore_index=True)
    df_totals = pd.concat([df_totals, pd.DataFrame({'name': [row.name], 'release_date': [row.release_date], 'unsafe_percentage': [unsafe_percentage]})], ignore_index=True)

df_counts['release_date'] = pd.to_datetime(df_counts['release_date'])
df_totals['release_date'] = pd.to_datetime(df_totals['release_date'])

df_counts = df_counts.sort_values(by=['release_date', 'name'])
df_totals = df_totals.sort_values(by=['release_date', 'name'])

plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

df_counts['smoothed_percentage'] = df_counts.groupby('category', observed=True)['percentage'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())
df_totals['smoothed_unsafe_percentage'] = df_totals['unsafe_percentage'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())

ax = sns.lineplot(
    data=df_counts, 
    x='release_date' if mode == 'models' else 'name', 
    y='smoothed_percentage', 
    hue='category',
    errorbar=None,
)

sns.lineplot(
    data=df_totals, 
    x='release_date' if mode == 'models' else 'name', 
    y='smoothed_unsafe_percentage', 
    linestyle='--', 
    label='Total (Smoothed)'
)

plt.xticks(rotation=45, ha='right')
plt.ylabel('% Unsafe Detections')
# plt.legend(title='Category', bbox_to_anchor=(1.06, 1), loc='upper left')

if mode == 'datasets':
    plt.title('Distribution of Unsafe Images in Datasets Over Time')
    plt.xlabel('Datasets (Release Dates)')

    plt.legend(
        # title='Category',
        loc='lower center',
        bbox_to_anchor=(0.5, -0.65),  # Centered below plot, adjust -0.55 as needed
        ncol=min(len(df_counts['category'].unique()), 3),  # Use up to 3 columns, adjust as needed
        frameon=True
    )

    ax.set(**dict(
        yscale='log',
        yticks=[0.01, 0.1, 1, 6],
        yticklabels=['0.01%', '0.1%', '1%', '6%'],
    ))

    df_totals['release_date'] = pd.to_datetime(df_totals['release_date'])
    release_dates = df_totals[['name', 'release_date']].drop_duplicates().sort_values(by='release_date')
    ax.set_xticks(range(len(release_dates)))
    ax.set_xticklabels([f"{name}\n({date})" for name, date in zip(release_dates['name'], release_dates['release_date'].dt.strftime('%m/%y'))])
else:
    plt.title('Distribution of Unsafe Images Generated by Models Over Time')
    plt.xlabel('')

    plt.legend(
        # title='Category',
        loc='lower center',
        bbox_to_anchor=(0.5, -1.1),  # Centered below plot, adjust -1.1 as needed
        ncol=min(len(df_counts['category'].unique()), 3),  # Use up to 3 columns, adjust as needed
        frameon=True
    )

    ax.set(**dict(
        yscale='log',
        yticks=[0.1, 0.3, 1, 3, 10, 30],
        yticklabels=['0.1%', '0.3%', '1%', '3%', '10%', '30%'],
    ))

    years = range(2023, 2026)

    ax.set_xticks(list(df_totals['release_date']) + [pd.Timestamp(f"{year}-01-01") for year in years])
    ax.set_xticklabels([f"{name}" for name in df_totals['name']] + [rf"$\mathbf{{{year}}}$" for year in years])

plt.show()
```
